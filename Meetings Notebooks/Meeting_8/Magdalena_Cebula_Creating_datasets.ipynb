```
import pandas as pd
import numpy as np

from imblearn.over_sampling import SMOTE, ADASYN 
from collections import Counter

from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

from sklearn.metrics import f1_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score

from sklearn.model_selection import train_test_split

from imblearn.under_sampling import RandomUnderSampler
from imblearn.over_sampling import RandomOverSampler
from imblearn.combine import SMOTETomek
from imblearn.under_sampling import NearMiss
from imblearn.over_sampling import ADASYN
```

```
dataset = pd.read_csv('/content/drive/My Drive/Colab Notebooks/Cancellation/data/noshow_processed_data_v2_Janek.csv', index_col = 'Unnamed: 0')
dataset.head()
```

```
Y = dataset["NoShow"]
Y.tail()
```

```
X = dataset.loc[:, dataset.columns != 'NoShow']
X.tail()
```

```
x, x_val, y, y_val = train_test_split(X, Y, test_size=0.1, train_size=0.9, random_state=0, stratify=Y, shuffle=True)
```

```
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.22, train_size =0.78, random_state=0, stratify=y, shuffle=True)
```

```
Train = pd.DataFrame(data = x_train, columns = X.columns)
YTrain = pd.DataFrame(data = y_train, columns = ["NoShow"])
Train["NoShow"] = YTrain
print(Train)

Train.to_csv('/content/drive/My Drive/Colab Notebooks/Cancellation/data/datasets/Train_raw.csv', index=False)
```

```
Test = pd.DataFrame(data = x_test, columns = X.columns)
YTest = pd.DataFrame(data = y_test, columns = ["NoShow"])
Test["NoShow"] = YTest
print(Test)

Test.to_csv('/content/drive/My Drive/Colab Notebooks/Cancellation/data/datasets/Test_raw.csv', index=False)
```

```
Val = pd.DataFrame(data = x_val, columns = X.columns)
YVal = pd.DataFrame(data = y_val, columns = ["NoShow"])
Val["NoShow"] = YVal
Val

Val.to_csv('/content/drive/My Drive/Colab Notebooks/Cancellation/data/datasets/Validation.csv', index=False)
```
